```python
### ESSE FUNCIONOU!!!!

import smtplib, ssl

port = 587  # For starttls
smtp_server = "smtp.gmail.com"
sender_email = "mgckdata@gmail.com"
receiver_email = "marcelgckaled@gmail.com"
password = input("Type your password and press enter:")
message = """\
Subject: Hi there

This message is sent from Python."""

context = ssl.create_default_context()
with smtplib.SMTP(smtp_server, port) as server:
    server.ehlo()  # Can be omitted
    server.starttls(context=context)
    server.ehlo()  # Can be omitted
    server.login(sender_email, password)
    server.sendmail(sender_email, receiver_email, message)
```

```python
#### ESSE FUNCIONOU COM HTML E ANEXO.

import email, smtplib, ssl

from email import encoders
from email.mime.base import MIMEBase
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

subject = f"Report Dados Covid-19 ({ontem})"
body = "Este é um email enviado pelo Python"
sender_email = "mgckdata@gmail.com"
receiver_email = "marcelgckaled@gmail.com"
password = input("Digite seu password e pressione Enter:")

# Create a multipart message and set headers
message = MIMEMultipart()
message["From"] = sender_email
message["To"] = receiver_email
message["Subject"] = subject
# message["Bcc"] = receiver_email  # Recommended for mass emails

# Create the plain-text and HTML version of your message
text = f"""\

Data da última coleta dos dados: {ontem}.



Este email foi enviado de forma automática. Favor não responder.
.
"""
html = """\
<html>
  <body>
    <p>Olá,<br>
       Como você está?<br>
       Segue report Covid-19 atualizado em D-1.
    </p>
  </body>
</html>
"""

# Turn these into plain/html MIMEText objects
part1 = MIMEText(text, "plain")
#part2 = MIMEText(html, "html")

# Add HTML/plain-text parts to MIMEMultipart message
# The email client will try to render the last part first
message.attach(part1)
#message.attach(part2)

filename = "evol_c19_cid.png" # In same directory as script

# Open PDF file in binary mode
with open(filename, "rb") as attachment:
    # Add file as application/octet-stream
    # Email client can usually download this automatically as attachment
    part = MIMEBase("application", "octet-stream")
    part.set_payload(attachment.read())

# Encode file in ASCII characters to send by email    
encoders.encode_base64(part)

# Add header as key/value pair to attachment part
part.add_header(
    "Content-Disposition",
    f"attachment; filename= {filename}",
)

# Add attachment to message and convert message to string
message.attach(part)
text = message.as_string()

# Log in to server using secure context and send email
context = ssl.create_default_context()
with smtplib.SMTP_SSL("smtp.gmail.com", 465, context=context) as server:
    server.login(sender_email, password)
    server.sendmail(sender_email, receiver_email, text)

print('Email enviado com sucesso!')
```

```python
#### ESSE FUNCIONOU COM HTML E ANEXO.

import email, smtplib, ssl

from email import encoders
from email.mime.base import MIMEBase
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

subject = f"Report Dados Covid-19 ({ontem})"
body = "Este é um email enviado pelo Python"
sender_email = "mgckdata@gmail.com"
receiver_email = "carloscso239@gmail.com"
password = '794613D@ta'

# Create a multipart message and set headers
message = MIMEMultipart()
message["From"] = sender_email
message["To"] = receiver_email
message["Subject"] = subject
#message["Bcc"] = receiver_email  # Recommended for mass emails

# Create the plain-text and HTML version of your message
text = f"""\

Data da última coleta dos dados: {ontem}.


Este email foi enviado de forma automática através de um programa Python. Favor não responder.

"""
html = """\
<html>
  <body>
    <p>Olá,<br>
       Como você está?<br>
       Segue report Covid-19 atualizado em D-1.
    </p>
  </body>
</html>
"""

# Turn these into plain/html MIMEText objects
part1 = MIMEText(text, "plain")
#part2 = MIMEText(html, "html")

# Add HTML/plain-text parts to MIMEMultipart message
# The email client will try to render the last part first
message.attach(part1)
#message.attach(part2)

files = ["evol_c19_cid.png",
         "log_estado.txt",
         "log_cidade.txt", 
         "log_paises.txt", 
         "log_continente.txt"
]
# In same directory as script

for f in files:

    # Open PDF file in binary mode
    with open(f, "rb") as attachment:
        # Add file as application/octet-stream
        # Email client can usually download this automatically as attachment
        part = MIMEBase("application", "octet-stream")
        part.set_payload(attachment.read())

    # Encode file in ASCII characters to send by email    
    encoders.encode_base64(part)

    # Add header as key/value pair to attachment part
    part.add_header("Content-Disposition",f"attachment; filename= {f}",)

    # Add attachment to message and convert message to string
    message.attach(part)
    text = message.as_string()

# Log in to server using secure context and send email
context = ssl.create_default_context()
with smtplib.SMTP_SSL("smtp.gmail.com", 465, context=context) as server:
    server.login(sender_email, password)
    server.sendmail(sender_email, receiver_email, text)

print('Email enviado com sucesso!')
```
