```python
import requests

download_url = "https://opendata.ecdc.europa.eu/covid19/casedistribution/csv"
target_csv_path = "covid19_worldwise_cases.csv"

response = requests.get(download_url)
response.raise_for_status()    # Confere se a requisição foi realizado com sucesso.
with open(target_csv_path, "wb") as f:
    f.write(response.content)
print("Download pronto.")
```

```python
import pandas as pd


df = pd.read_csv('covid19_worldwise_cases.csv')
pd.set_option('display.max_rows',None)
df.info()
```

```python
df_brazil = df[df['countriesAndTerritories'] == 'Brazil']
df_brazil
```

```python
# data do primeiro caso = 26/02/2020
```

```python

def pandas_test():
    return df_brazil['cases'].sum()

def numpy_test():
    return df_brazil['cases'].to_numpy().sum()

print(timeit.timeit(numpy_test, number=)) # 0.5032469799989485
print(timeit.timeit(pandas_test, number=))  # 0.6035906639990571

print(df_brazil['cases'].sum())
print(df_brazil['deaths'].sum())
```

```python

stdoutOrigin=sys.stdout 
sys.stdout = open("log_paises.txt", "w")

print('Dados Covid-19 por país:')
print()
print(r'{casos confirmados(c) | mortes(m)}')
print()
for i in df.countriesAndTerritories.unique():
    a = df['cases'][df['countriesAndTerritories'] == i].sum()
    b = df['deaths'][df['countriesAndTerritories'] == i].sum()
    print(f'{i}: {a} | {b}')
    
```

```python

stdoutOrigin=sys.stdout 
sys.stdout = open("log_continente.txt", "w")

print('Dados Covid-19 por continente:')
print()
print(r'{casos confirmados(%) | mortes(%)}')
print()
for cont in df.continentExp.unique():
    c = df['cases'][df['continentExp'] == cont].sum()
    d = df['deaths'][df['continentExp'] == cont].sum()
    e = (df['cases'][df['continentExp'] == cont].sum() / df['cases'].sum()) * 100
    f = (df['deaths'][df['continentExp'] == cont].sum() / df['deaths'].sum()) * 100
    
    print(f'{cont}: {c}({e:.2f}%) | {d}({f:.2f}%)')

sys.stdout.close()
sys.stdout=stdoutOrigin
```
